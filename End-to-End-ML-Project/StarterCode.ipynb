```python
# Step 0: Import libraries
from sklearn.datasets import fetch_california_housing
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the dataset
housing = fetch_california_housing()
housing_df = pd.DataFrame(housing.data, columns=housing.feature_names) # create a Pandas Dataframe, a two dimensional data structure similar to a table.
housing_df.head() # # This returns the first 5 rows in our data frame.
```

```python
# Step 2: Simulate missing values (practice preprocessing)
# TODO: Make some values in one column missing
# Example: housing_df.loc[0:4, 'AveRooms'] = None
housing_df.loc[0:4, 'HouseAge'] = None # Using .loc to modify rows/columns by their name, not index. In this case, I am removing HouseAge from rows 0-4.
housing_df.head()
```

```python
# Step 3: Fill missing values
# TODO: Fill missing values with the median of the column
# Example: housing_df['AveRooms'] = housing_df['AveRooms'].fillna(housing_df['AveRooms'].median())
housing_df['HouseAge'] = housing_df['HouseAge'].fillna(housing_df['HouseAge'].median()) # selecting the HouseAge column and setting it to fill missing values (.fillna) with the median value of the House Age column
housing_df.head()
```

```python
# Step 4: Explore the data with a scatter plot
# TODO: Create a scatter plot of Longitude vs Latitude
# Color the points using housing.target
# Hint: plt.scatter(x, y, c=..., cmap='viridis', alpha=0.5)

plt.figure(figsize=(10, 6)) # Set the size of the figure
plt.scatter(
    housing_df['Longitude'], # x axis
    housing_df['Latitude'], # y axis
    c=housing.target, # Color of points based on median house value
    cmap='inferno', # color map
    alpha=0.5 # transparency
)
plt.colorbar(label='Median House Value ($100,000)')
plt.title('California Housing: Latitude vs. Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
