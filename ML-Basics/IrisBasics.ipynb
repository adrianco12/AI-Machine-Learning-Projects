```python
from sklearn.datasets import load_iris # importing the iris dataset
import matplotlib.pyplot as plt # importing matplotlib, which will be used for our histogram
iris = load_iris() # loading iris into a variable

selected_features = [iris.feature_names[2], iris.feature_names[3]] # Getting specific features (Petal length, width)
print("Feature names:", selected_features)
print("Target names:", iris.target_names[2])  # Showing a specific species of flower
virginica_data = iris.data[iris.target == 2] # Storing Virginica data in a variable

plt.hist(virginica_data[:, 2], bins=10, label='Petal Length') # Adding petal length to the histogram
plt.hist(virginica_data[:, 3], bins=10, label='Petal Width') # Adding petal width to the histogram
plt.title("Histogram of Petal Size for Virginica")
plt.xlabel("Size (cm)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
